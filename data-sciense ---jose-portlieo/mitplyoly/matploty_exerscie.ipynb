```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0,100)
y=x*2
z=x**2
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,0.2,0.2])
# axes1.plot(x,y,colors="#ff8C00")
# axes2.plot(x,y,colors="#ff8C00")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.8,0.2,0.2])
axes1.plot(x,z)
axes2.plot(x,z)
axes1.set_ylim(2000,10000)
axes2.set_ylim(100,500)
axes1.set_xlim(0,100)
axes2.set_xlim(20,22)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot()
axes[1].plot()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(15,5))
axes[0].plot(x,y, lw=3,ls="--")
axes[1].plot(x,z, lw=3, color="orange")
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2 , figsize=(15,5))
axes[0].plot(x,y, lw=3,color="orange",ls="-")
axes[1].plot(x,z, lw=3, ls="--")
```
