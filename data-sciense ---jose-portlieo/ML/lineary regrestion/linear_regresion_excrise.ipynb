```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("D:/Learning Stuff/python/data-sciense ---jose-portlieo/ML/lineary regrestion/Ecommerce Customers.csv")
df.head()
```

```python
x = df[['Avg. Session Length','Time on App','Time on Website','Length of Membership']]
x
```

```python
y = df[['Yearly Amount Spent']]
y
```

```python
df.describe()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=101,train_size=0.4)
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
lr.coef_[0]
```

```python
x_predition = lr.predict(x_test)
x_predition
```

```python
plt.scatter(y_test,x_predition)
```

```python
sns.distplot((y_test-x_predition))
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
mae = mean_absolute_error(y_test, x_predition)
mse = mean_squared_error(y_test, x_predition)
rmse = np.sqrt(mse)
mae,mse,rmse
```

```python
cdf = pd.DataFrame(lr.coef_[0],x.columns,columns=['coeff'])
cdf 
```
